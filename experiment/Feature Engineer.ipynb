# Feature Engineer

# 1. Load Data

```python
import pandas as pd
import numpy as np

df = pd.read_csv("evaldata_cleaned.csv")

df.head()
```

# 2. Drop Identifiers

```python
df = df.drop(columns=["ANON_ID"], errors="ignore")
```

# 3. Construct Feature

```python
df["Birthdate"] = pd.to_datetime(df["Birthdate"], errors="coerce")

df["age"] = df["year"] % 100 + 2000 - df["Birthdate"].dt.year
```

```python
df["absence_rate"] = df["DaysAbs"] / df["DaysEnr"]
```

```python
df["has_suspension"] = (df["Susp"] > 0).astype(int)
```

```python
def school_stage(g):

    if g <= 5:
        return "elementary"
    elif g <= 8:
        return "middle"
    else:
        return "high"

df["school_stage"] = df["Grade"].apply(school_stage)
```

```python
df["crime_rate"] = df["total_crimes"] / df["total_population"]
df["violent_crime_rate"] = df["violent_crimes"] / df["total_population"]
df["property_crime_rate"] = df["property_crimes"] / df["total_population"]
df["drug_crime_rate"] = df["drug_crimes"] / df["total_population"]
df["other_crime_rate"] = df["other_crimes"] / df["total_population"]
```

Interaction Features

```python
df["poverty_crime"] = df["poverty_rate_pct"] * df["crime_rate"]
```

```python
df["gpa_absence"] = df["CurrWeightedTotGPA"] * df["absence_rate"]
```

# 4. Save Engineered Dataset

```python
df.to_csv("engineered_features.csv", index=False)

print("Feature engineering complete.")
print("Saved file: engineered_features.csv")
```
